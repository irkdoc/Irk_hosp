Сделаем колонку с активными случаями.
В статистике "стопкоронавирус.рф" это расчетный показатель, вычисляемый по формуле:

A = C - (R + D),

где A — активные случаи, C — подтвержденные диагнозы (cases), R — выздоровевшие (cured), D — умершие(death). 

```python
df_yand['active']=df_yand['cases']-df_yand['cured']-df_yand['death']
```

Сделаем колонку с заболеваемостью по дням (delta) на 100 000 населения (popul)

```python
df_yand['cases_100k_delta']=100000*df_yand['cases_delta']/df_yand['popul']
```

Сделаем колонку с летальностью (CFR- case fatality rate) death/cases
и колонку с летальностью по дням (cfr_delta) death_delta/cases_delta

```python
df_yand['cfr']=df_yand['death']/df_yand['cases']
df_yand['cfr_delta']=df_yand['death_delta']/df_yand['cases_delta']
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import correlate

a,b, N = 0, 10, 1000        #Boundaries, datapoints
shift = -0.2                  #Shift, note 3/10 of L = b-a

x = np.linspace(a,b,N)
x1 = 1*x + shift
time = np.arange(1-N,N)     #Theoritical definition, time is centered at 0

y1 = sum([np.sin(2*np.pi*i*x/b) for i in range(1,5)])
y2 = sum([np.sin(2*np.pi*i*x1/b) for i in range(1,5)])

#Really only helps with large irregular data, try it
# y1 -= y1.mean()
# y2 -= y2.mean()
# y1 /= y1.std()
# y2 /= y2.std()

# cross_correlation = correlate(y1,y2)
cross_correlation = correlate(y1,y2, method='fft')
print (cross_correlation.argmax())
print (time[cross_correlation.argmax()])
shift_calculated = time[cross_correlation.argmax()] *1.0* b/N
y3 = sum([np.sin(2*np.pi*i*(x1-shift_calculated)/b) for i in range(1,5)])+1
print ("Preset shift: ", shift, "\nCalculated shift: ", shift_calculated)



plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.legend(("Regular", "Shifted", "Recovered"))
plt.savefig("SO_timeshift.png")
plt.show()
```

```python
import numpy as np

X = np.linspace(0,2*np.pi,30)  #some X values

def yvals(x):
    return np.sin(x)+np.sin(2*x)+np.sin(3*x)

Y1 = yvals(X)
Y2 = yvals(X-0.1)  #shifted y values

#fourier transform both series
FT1 = np.fft.fft(Y1)
FT2 = np.fft.fft(Y2)

#You can show that analyically, a phase shift in the coefficients leads to a 
#multiplicative factor of `exp(-1.j * N * T_d)`

#can't take the 0'th element because that's a division by 0.  Analytically, 
#the division by 0 is OK by L'hopital's<sp?> rule, but computers don't know calculus :)
print (np.log(FT2[1:]/FT1[1:])/(-1.j*np.arange(1,len(X))))
```
